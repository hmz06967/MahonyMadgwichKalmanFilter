```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# MADGWICH: Read The Filter Response Data

```python
data = pd.read_csv("mg_mmkf.log", header=None, delimiter=" ",encoding="utf-16-le")
sample_z = np.array(data[0])
z_mmkf_mg = np.array(data[1])
v_mmkf_mg = np.array(data[3])
pvv_mmkf_mg = np.array(data[4])
pzz_mmkf_mg = np.array(data[2])
```

```python
data = pd.read_csv("mg_mmkf_eul.log", header=None, delimiter=" ",encoding="utf-16-le")
mg_yaw = data[0]
mg_pitch = data[1]
mg_roll = data[2]
```

# MAHONY: Read The Filter Response Data

```python
data = pd.read_csv("mmkf.log", header=None, delimiter=" ",encoding="utf-16-le")
z_mmkf_mh = np.array(data[1])
v_mmkf_mh = np.array(data[3])
pvv_mmkf_mh = np.array(data[4])
pzz_mmkf_mh = np.array(data[2])
```

```python
data = pd.read_csv("mmkf_eul.log", header=None, delimiter=" ",encoding="utf-16-le")
mh_yaw = data[0]
mh_pitch = data[1]
mh_roll = data[2]
```

```python
# enlarge the plot size to show more detail
plt.rcParams["figure.figsize"]=50,10

#  barometric pressure sensor derived altitude samples
plt.plot(sample_z, color='yellow', label='Sample Altitude')
plt.plot(z_mmkf_mg, color='blue', label='Mahny Estimate Altitude')
plt.plot(z_mmkf_mh, color='green', label='Madg Estimate Altitude')
#plt.plot(z_mmkf, color='red', label='Madg Estimate Altitude')

plt.title('Kalman Filter Test')
plt.legend()
plt.xlabel('Time sample')
plt.ylabel('Altitude cm')

plt.show()
```

```python
# enlarge the plot size to show more detail
plt.rcParams["figure.figsize"]=30,10
plt.rcParams["figure.facecolor"]="w"

#  barometric pressure sensor derived altitude samples
plt.plot(v_mmkf_mg, color='blue', label='Mahny Velocity ')
plt.plot(v_mmkf_mh, color='green', label='Madg Velocity ')

plt.title('Sensr data')
plt.legend()
plt.xlabel('Time sample')
plt.ylabel('Velocity (cm/s)')

plt.show()
```

```python
# enlarge the plot size to show more detail
plt.rcParams["figure.figsize"]=30,10
plt.rcParams["figure.facecolor"]="w"

#  barometric pressure sensor derived altitude samples
plt.plot(pzz_mmkf_mg, color='blue', label='Mahny Altitude  Uncertainty')
plt.plot(pzz_mmkf_mh, color='green', label='Madg Altitude  Uncertainty')
plt.title('Altitude Uncertainty Pzz')
plt.xlabel('Time sample')
plt.ylabel('Pzz (cm^2)')
plt.legend()
plt.show()
```

```python
# enlarge the plot size to show more detail
plt.rcParams["figure.figsize"]=30,10
plt.rcParams["figure.facecolor"]="w"

#  barometric pressure sensor derived altitude samples
plt.plot(pvv_mmkf_mg, color='blue', label='Mahny Velocity Uncertainty')
plt.plot(pvv_mmkf_mh, color='green', label='Madg Velocity Uncertainty')

plt.title('Sensr data')
plt.legend()
plt.xlabel('Time sample')
plt.ylabel('Pvv (cm/s)^2')
plt.title('Velocity uncertainty Pvv')

plt.show()
```

```python
# enlarge the plot size to show more detail
plt.rcParams["figure.figsize"]=30,10
plt.rcParams["figure.facecolor"]="w"

d2r = 0.0174533
#  barometric pressure sensor derived altitude samples
plt.plot(mg_yaw[0:100], color='red', label='yaw')
plt.plot(mg_pitch[0:100], color='blue', label='pitch')
plt.plot(mg_roll[0:100], color='green', label='roll')

#  barometric pressure sensor derived altitude samples
plt.plot(d2r*mh_yaw[0:100], color='brown', label='mh yaw')
plt.plot(d2r*mh_pitch[0:100], color='purple', label='mh pitch')
plt.plot(d2r*mh_roll[0:100], color='black', label='mh roll')


plt.title('Sensr data')
plt.legend()
plt.xlabel('Time sample')
plt.ylabel('Angular Rotation')
plt.title('rad/s')

plt.show()
```
